## Q1 — Adicionar coluna `faturamento`

Adicione a coluna `faturamento` (`qtd * preco`) **sem modificar** o DataFrame original.  
Use `.assign()` — abordagem imutável.

````python
# Q1 - Adicionar coluna faturamento
# ✅ Use lambda dentro do .assign() — funciona corretamente em method chains
vendas_faturadas = vendas.assign(faturamento=lambda df: df["qtd"] * df["preco"])
vendas_faturadas
````

### 📝 Q1 — Revisão

**Nota: 9/10**

Lógica correta. Ponto de atenção: usar `lambda df:` dentro do `.assign()` é o padrão robusto.

```python
# ❌ Referência externa — quebra se .assign() estiver no meio de uma cadeia
vendas.assign(faturamento=vendas["qtd"] * vendas["preco"])

# ✅ Lambda recebe o df no estado atual da cadeia
vendas.assign(faturamento=lambda df: df["qtd"] * df["preco"])
```

**Por quê:** quando você encadeia `.dropna().rename().assign(...)`, o `df` recebido pelo lambda é o resultado *pós-rename*, não o `vendas` original. A referência externa aponta sempre para o objeto nomeado — e pode estar desatualizado.

## Q2 — Filtrar pedidos relevantes

Filtre apenas os pedidos com `status == "entregue"` **e** `faturamento > 500`.  
Armazene o resultado em `vendas_relevantes`.

````python
# Q2 - Filtrar pedidos relevantes
# Alternativa limpa usando o método .query():

vendas_relevantes = vendas_faturadas.query("status == 'entregue' and faturamento > 500")
vendas_relevantes
````

### 📝 Q2 — Revisão

**Nota: 10/10**

`.query()` é a escolha idiomática. Vantagens vs boolean mask:

| | Boolean mask | `.query()` |
|---|---|---|
| Legibilidade | `df[(df["a"] > 0) & (df["b"] == "x")]` | `df.query("a > 0 and b == 'x'")` |
| Variáveis Python | `df[df["a"] > limite]` | `df.query("a > @limite")` — prefixo `@` |
| Performance | padrão | usa `numexpr` se instalado (mais rápido em > 100k linhas) |
| Colunas com espaço | — | use backticks: `` `nome coluna` `` |

Os 5 pedidos restantes são exatamente os esperados: Mouse (R$269.70) e Teclado (R$300.00) foram removidos por `faturamento ≤ 500`, e o Headset foi removido por `status != 'entregue'`.

## Q3 — Resumo por vendedor

Usando `groupby` em `vendas_relevantes`, calcule por **vendedor**:
- `total_pedidos` — quantidade de pedidos
- `faturamento_total` — soma do faturamento
- `ticket_medio` — média do faturamento

Ordene por `faturamento_total` decrescente.

````python
# Q3 - Resumo por vendedor
resumo_vendedor = vendas_relevantes.groupby('vendedor').agg(
    total_pedidos=('pedido', 'count'),
    faturamento_total=('faturamento', 'sum'),
    ticket_medio=('faturamento', 'mean')
).sort_values(by='faturamento_total', ascending=False)
resumo_vendedor
````

### 📝 Q3 — Revisão

**Nota: 10/10**

Named aggregations com tuplas `("coluna_fonte", "função")` é o padrão de produção — cria nomes de colunas limpos direto no `.agg()`.

**Comportamento do índice pós-groupby:**
```python
# Após .groupby().agg(), o campo agrupado vira o INDEX
resumo_vendedor.index   # Index(["Ana L", "Bruno", "Diego"], name="vendedor")

# Para manter como coluna regular:
.groupby("vendedor", as_index=False).agg(...)
# ou:
.reset_index()
```

**Resultado esperado:**
```
vendedor    total_pedidos  faturamento_total   ticket_medio
Bruno               3           8060.0          2686.67
Ana L               1           5600.0          5600.00
Diego               1           1140.0          1140.00
```

## Q4 — Participação por categoria

Usando `groupby` em `vendas_relevantes`, calcule o `faturamento_total` por **categoria**.  
Depois adicione uma coluna `percentual` com a participação de cada categoria no total geral.

````python
# Q4 - Participação por categoria
# ✅ Lambda no .assign() — recebe o df resultante do .agg(), não o externo
resumo_categoria = (
    vendas_relevantes
    .groupby("categoria")
    .agg(faturamento_total=("faturamento", "sum"))
    .assign(
        percentual=lambda df: df["faturamento_total"] / df["faturamento_total"].sum() * 100
    )
)
resumo_categoria
````

### 📝 Q4 — Revisão

**Nota: 9/10** (corrigido para 10/10 com o `lambda`)

**Por que o `lambda` é essencial aqui:** o `.assign()` está encadeado após `.agg()`. O `lambda df:` recebe o DataFrame resultante do `.agg()` — que é diferente de `vendas_relevantes`. Sem lambda, a referência ao nome externo apontaria para o objeto errado.

**Alternativa com `.transform()` — mantém uma linha por pedido (útil para análise por linha):**
```python
# .transform() é a "window function" do pandas — agrega mas não colapsa o shape
vendas_relevantes = vendas_relevantes.assign(
    total_cat  =lambda x: x.groupby("categoria")["faturamento"].transform("sum"),
    pct_cat    =lambda x: x["faturamento"] / x["total_cat"] * 100,
)
# Análogo SQL: SUM(faturamento) OVER (PARTITION BY categoria)
```

**Resultado esperado:**
```

categoria

Eletrônicos             12900.0       87.46

```

## Q5 — Produto de maior faturamento por cliente

Para cada **cliente**, encontre o produto com maior faturamento individual (pedido unitário).  
Use `groupby` + `idxmax()` ou qualquer método sem loop `for`.

````python
# Q5 - Produto de maior faturamento por cliente

# 1. Pegamos os índices das linhas onde ocorre o maior faturamento para cada cliente
idx_top_produtos = vendas_faturadas.groupby('cliente')['faturamento'].idxmax()

# 2. Filtramos o DataFrame original apenas com esses índices, trazendo colunas específicas
top_produtos_cliente = vendas_faturadas.loc[
    idx_top_produtos, 
    ['cliente', 'produto', 'faturamento']
]
top_produtos_cliente
````

### 📝 Q5 — Revisão

**Nota: 10/10**

Padrão correto e eficiente. Entender o que cada etapa retorna é o ponto central:

```python
# Etapa 1: .idxmax() por grupo → retorna os ÍNDICES de linha, não os valores
idx_top = vendas_faturadas.groupby("cliente")["faturamento"].idxmax()
# cliente
# Ana        ...   → índice da linha com maior faturamento para Ana
# Beatriz    ...
# João       ...
# dtype: int64

# Etapa 2: .loc[] usa esses índices para buscar as linhas no df original
vendas_faturadas.loc[idx_top, ["cliente", "produto", "faturamento"]]
```

**Decisão de escopo:** `vendas_faturadas` inclui pedidos cancelados/pendentes. Em produção, documente a escolha:
```python
# Opção A: histórico completo (inclui cancelados)
base = vendas_faturadas

# Opção B: apenas pedidos concluídos
base = vendas_faturadas.query("status == 'entregue'")

idx_top = base.groupby("cliente")["faturamento"].idxmax()
```

**Alternativa com `sort_values + drop_duplicates` — mais legível para alguns:**
```python
(
    vendas_faturadas
    .sort_values("faturamento", ascending=False)
    .drop_duplicates(subset="cliente", keep="first")
    [["cliente", "produto", "faturamento"]]
    .sort_values("cliente")
)
# Mesma lógica, diferente estilo — idxmax() é mais eficiente (O(n) vs O(n log n))
```
